I copied and pasted the list of fixtures from: https://draft.premierleague.com/league into a spreadsheet program and downloaded it as a csv. I will now clean the data and add it as a table in ym database

```python
import pandas as pd
import pymysql
```

```python
df = pd.read_csv('fixtures - Sheet1.csv')
```

```python
df
```

the patern seems to be a string indicating the gameweek followed by three fixtures. I will parse the gameweek from the string and create a list with the gameweek for each row

```python
gameweek = []
indexes = []

for i in range(len(df)):
    li = df['player_1'].iloc[i].split()
    if li[0] == 'Gameweek':
        # appending the index of the rows with the gameweek string to drop them
        indexes.append(i)
        for i in range(3):
            gameweek.append(li[1])
```

```python
gameweek
```

```python
# Now that I have a list with the gameweeks and the indexes of the rows containing the gameweek string
# I will drop the unwanted rows and add the column noting the gameweek per row

df.drop(indexes,inplace=True)
df['gameweek']=gameweek
```

```python
len(gameweek)
```

```python
df
```

```python
# I only want this table to contain the fixtures so I will drop the scores of any fixtures which have alread taken place.
df.drop('scores',axis=1,inplace=True)
```

```python
df
```

```python
# I also only want to team name not the player name so I will remove the player names from the columns

player_1=[]
player_2=[]
for i in range(len(df)):
    pl1_list = df['player_1'].iloc[i].split('\n') 
    pl2_list = df['player_2'].iloc[i].split('\n')
    player_1.append(pl1_list[0])
    player_2.append(pl2_list[0])
    
df['player_1'] = player_1
df['player_2'] = player_2
```

```python
df
```

```python
# I will now create a connection to my db I have already created in the GUI

connection = pymysql.connect(host='localhost',
                             user='root',
                             password='password',
                             db='fantasy_football')

cursor = connection.cursor()
```

```python
# Now I will iterate over my dataframe and add it to the table I have created in the GUI

sql = "INSERT INTO `fixtures` (`player_1`, `player_2`, `gameweek`) VALUES (%s, %s, %s)"

for i in range(len(df)):
    player_1 = df['player_1'].iloc[i]
    player_2 = df['player_2'].iloc[i]
    gameweek = df['gameweek'].iloc[i]
    cursor.execute(sql,(player_1,player_2 ,gameweek))
    connection.commit()
```
